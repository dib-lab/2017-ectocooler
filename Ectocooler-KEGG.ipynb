```python
import pyupset as pyu
import matplotlib as mpl
import matplotlib.pyplot as plt
from pickle import load
import pandas as pd
import glob 

%matplotlib inline
```

```python
#Import all of the data

pIDkID_Hash={}
for file in glob.glob('*ko'):
    df=pd.read_table(file, names=['pID','kID'])
    name=file.split('.')[0]
    pIDkID_Hash[name]=df
```

```python
grouped_hash={}
for ec in pIDkID_Hash.keys():
    df=pIDkID_Hash[ec]
    df=df.groupby('kID').count()
    df=df.reset_index()
    grouped_hash[ec]=df
```

```python
sub_hash=grouped_hash.copy()
del_list=['t_sp47', 't_ovolyt','t_maritim']
for i in del_list:
    del sub_hash[i]
```
